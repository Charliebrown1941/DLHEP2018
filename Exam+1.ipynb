```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

Names = "EventId,DER_mass_MMC,DER_mass_transverse_met_lep,DER_mass_vis,DER_pt_h,DER_deltaeta_jet_jet,DER_mass_jet_jet,DER_prodeta_jet_jet,DER_deltar_tau_lep,DER_pt_tot,DER_sum_pt,DER_pt_ratio_lep_tau,DER_met_phi_centrality,DER_lep_eta_centrality,PRI_tau_pt,PRI_tau_eta,PRI_tau_phi,PRI_lep_pt,PRI_lep_eta,PRI_lep_phi,PRI_met,PRI_met_phi,PRI_met_sumet,PRI_jet_num,PRI_jet_leading_pt,PRI_jet_leading_eta,PRI_jet_leading_phi,PRI_jet_subleading_pt,PRI_jet_subleading_eta,PRI_jet_subleading_phi,PRI_jet_all_pt,Weight,Label"
Splitnames = Names.split(",")
print Splitnames

filename = "mytraining.csv"

df = pd.read_csv(filename, names=Splitnames)
df

```

```python
df_sig=df[df.Label== 's']
df_bkg=df[df.Label== 'b']

for var in Splitnames[:-1]:
    print var
    plt.figure()
    plt.hist(df_sig[var],bins=200,histtype="step", color="red",label="background",stacked=True)
    plt.hist(df_bkg[var],bins=200,histtype="step", color="blue", label="signal",stacked=True)
    plt.legend(loc='upper right')
    plt.show()
```

```python
import sklearn.discriminant_analysis as DA
Fisher=DA.LinearDiscriminantAnalysis()


N_Train=25000

Train_Sample=df[:N_Train]
Test_Sample=df[-N_Train:]

X_Train=Train_Sample[Splitnames[:-1]]
y_Train=Train_Sample["Label"]

X_Test=Test_Sample[Splitnames[:-1]]
y_Test=Test_Sample["Label"]

Test_sig=Test_Sample[Test_Sample.Label=='s']
Test_bkg=Test_Sample[Test_Sample.Label=='b']

Binarylist = [0 if x == 'b' else 1 for x in y_Test]


Fisher.fit(X_Train,y_Train)

plt.figure()
plt.hist(Fisher.decision_function(Test_sig[Splitnames[:-1]]),bins=100,histtype="step", color="blue", label="signal",stacked=True)
plt.hist(Fisher.decision_function(Test_bkg[Splitnames[:-1]]),bins=100,histtype="step", color="red", label="background",stacked=True)
plt.legend(loc='upper right')
plt.show()

```

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr, _ = roc_curve(Binarylist, Fisher.decision_function(X_Test))

roc_auc = auc(fpr, tpr)

plt.plot(fpr,tpr,color='black',label='ROC curve (area = %0.2f)' % roc_auc)
plt.legend(loc="lower right")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

plt.show()
```
